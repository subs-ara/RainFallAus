```python
#Import our data
df = pd.read_csv('weatherAUS.csv')
```

```python
#since columns like Sunshine, Evaporation, Cloud3pm, Cloud9am have most of its values as null, we shall eliminate those columns
# Owing to the context, Date , location are irrelevant for analysis
```

```python
df.drop(columns=['Evaporation', 'Sunshine', 'Cloud3pm', 'Cloud9am','Date', 'Location'],axis=1, inplace=True)
```

```python
df.head()
```

```python
df.info()
```

```python
df['RainToday'].replace({'No': 0, 'Yes': 1}, inplace = True)
df['RainTomorrow'].replace({'No': 0, 'Yes': 1}, inplace = True)
```

```python
df.head() #above we replaced No with 0 and Yes with 1
```

```python
#some features still need to be converted into numerical form
features_to_be_converted = ['WindGustDir', 'WindDir9am', 'WindDir3pm']
df_final = pd.get_dummies(df, columns=features_to_be_converted)
df_final.head()
```

```python
df_final.dropna(axis=0,inplace=True)
```

```python
#Standardisation / Scaling
```

```python
#Many ML algorithms perform better when features are on a relatively similar scale , or in other words normally distributed
```

```python
#We are going to use preprocessing.MinMaxScaler library to do the standardization
```

```python
scaler = preprocessing.MinMaxScaler()
scaler.fit(df_final)
df_final = pd.DataFrame(scaler.transform(df_final),index=df_final.index, columns=df_final.columns)
df_final.head()
```
